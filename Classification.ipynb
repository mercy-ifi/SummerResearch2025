# Data Preparation

```python
df = df.drop(['affilatns_ind',
       'affilatn_code_1', 'affilatn_code_2', 'affilatn_code_3',
       'affilatn_code_4', 'affilatn_code_5', 'affilatn_code_6',
       'affilatn_code_7', 'affilatn_code_8', 'affilatn_code_9',
       'affilatn_code_10'], axis=1)
```

Going to narrow down the data frame in order to look into classification/custody level and relationships with key demographics(race, sex, institution)

```python
df = df[['control_number', 'custody','race', 'sex', 'date_of_birth', 'commit_instit',
       'boot_camp_elig', 'program_level_1', 'program_level_2',
       'program_level_3', 'ic_user_id', 'ic_data_entry_dt', 'ic_lst_name',
       'ic_fst_initial', 'ic_title', 'ic_date', 
       'ic_custdy_level']]
```
